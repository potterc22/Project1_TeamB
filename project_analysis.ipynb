```python
# Import Dependencies
import matplotlib.pyplot as plt 
import os
import csv
import numpy as np 
import pandas as pd 
import pandas_profiling
#import altair as alt
```

```python
# Declare file path for nfl_teams.csv
teams_csv = os.path.join('Data_files', 'nfl_teams.csv')
# Create an empty list to hold the teams data, to help us create a Dataframe
teams_list = []
with open(teams_csv, 'r') as csvfile:
    # CSV reader specifies delimiter and variable that holds contents
    teams_data = csv.reader(csvfile, delimiter=',')
    # Read the header row first
    teams_header = next(teams_data)
    #print header row
    print(f'CSV Header: {teams_header}')
    # Read each row of data after the header
    for row in teams_data:
        # add each row of data to our teams list
        teams_list.append(row)
        print(row)
```

```python
teams_df = pd.DataFrame(teams_list, columns= teams_header)
teams_df.head()
```

```python
# Declare file path for stadiums.csv
stadiums_csv = os.path.join('Data_files', 'stadiums.csv')
# Create an empty list to hold the stadiums data, to help us create a Dataframe
stadiums_list = []
with open(stadiums_csv, 'r') as csvfile:
    # CSV reader specifies delimiter and variable that holds contents
    stadium_data = csv.reader(csvfile, delimiter=',')
    # Read the header row first
    stadium_header = next(stadium_data)
    #print header row
    print(f'CSV Header: {stadium_header}')
    # Read each row of data after the header
    for row in stadium_data:
        # add each row of data to our stadiums list
        stadiums_list.append(row)
        print(row)
```

```python
stadium_df = pd.DataFrame(stadiums_list, columns= stadium_header)
stadium_df.head()
```

```python
# Declare file path for spreadspoke_scores.csv
scores_csv = os.path.join('Data_files', 'spreadspoke_scores.csv')
# Create an empty list to hold the scores data, to help us create a Dataframe
scores_list = []
with open(scores_csv, 'r') as csvfile:
    # CSV reader specifies delimiter and variable that holds contents
    score_data = csv.reader(csvfile, delimiter=',')
    # Read the header row first
    score_header = next(score_data)
    #print header row
    print(f'CSV Header: {score_header}')
    # Read each row of data after the header
    for row in score_data:
        # add each row of data to our scores list
        scores_list.append(row)
        print(row)
```

```python
scores_df = pd.DataFrame(scores_list, columns= score_header)
scores_df = scores_df.rename(columns={'stadium':'stadium_name'})
scores_df.head()
```

```python
# convert schedule season column to numeric so we can filter for 2002-2018
scores_df['schedule_season'] = pd.to_numeric(scores_df['schedule_season'])
scores_2002_2018 = scores_df.loc[(scores_df['schedule_season'] >= 2002) & (scores_df['schedule_season'] <= 2018)]
scores_2002_2018.head()
```

```python
afc_teams = teams_df.loc[
    (teams_df['team_division'] == 'AFC North') | 
    (teams_df['team_division'] == 'AFC East') | 
    (teams_df['team_division'] == 'AFC South') | 
    (teams_df['team_division'] == 'AFC West'),:]
afc_teams
```

```python
nfc_teams = teams_df.loc[
    (teams_df['team_division'] == 'NFC North') | 
    (teams_df['team_division'] == 'NFC East') | 
    (teams_df['team_division'] == 'NFC South') | 
    (teams_df['team_division'] == 'NFC West'),:]
nfc_teams
```

```python
scores_stadium_merged = pd.merge(scores_df, stadium_df, on='stadium_name',how='left')
scores_stadium_merged['score_home'] = scores_stadium_merged['score_home'].astype(float)
scores_stadium_merged['score_away'] = scores_stadium_merged['score_away'].astype(float)
scores_stadium_merged['total_score'] = scores_stadium_merged['score_home'] + scores_stadium_merged['score_away']
scores_stadium_merged.head()
```

```python
merged_2002_2018 = pd.merge(scores_2002_2018, stadium_df, on='stadium_name',how='left')
# remove playoff data because there was a good amount of missing data
merged_2002_2018 = merged_2002_2018.loc[merged_2002_2018['schedule_playoff'] == 'FALSE']
# convert weather, home and away scores to float
merged_2002_2018['score_home'] = merged_2002_2018['score_home'].astype(float)
merged_2002_2018['score_away'] = merged_2002_2018['score_away'].astype(float)
merged_2002_2018['weather_temperature'] = pd.to_numeric(merged_2002_2018['weather_temperature'])
merged_2002_2018['over_under_line'] = pd.to_numeric(merged_2002_2018['over_under_line'])
# Add total score column
merged_2002_2018['total_score'] = merged_2002_2018['score_home'] + merged_2002_2018['score_away']
merged_2002_2018.head()
```

```python
teams = []
[teams.append(team) for team in afc_teams['team_name']]
[teams.append(team) for team in nfc_teams['team_name']]
teams.sort()
teams
```

```python
team_stats = pd.DataFrame(teams)
team_stats = team_stats.rename(columns={0: 'team_name'})
team_stats['total_points_scored'] = 0
team_stats['total_games_played'] = 0
team_stats['total_wins'] = 0
team_stats['total_losses'] = 0
team_stats['total_temp'] = 0
team_stats['home_wins'] = 0
team_stats['home_games'] = 0
team_stats.head()
```

```python
# Use count to index through our teams 
count = 0


# Loop through each team
for team in teams:
    # Loop through each row of the merged data for 2002-2018
    for index, row in merged_2002_2018.iterrows():
        # loop through each row and looking for team in home and adding the score to our dataframe 
        if row['team_home'] == team:
            team_stats.loc[count, 'total_points_scored'] += row['score_home']
            team_stats.loc[count, 'total_games_played'] += 1
            team_stats.loc[count, 'home_games'] += 1
            # Add each weather value that is not NaN to total temp to allow us to calculate avg temperature 
            if row['weather_temperature'] != 'NaN':
                team_stats.loc[count, 'total_temp'] += row['weather_temperature']
            else:
                continue
            # Calculate the number of wins and losses
            if row['score_home'] > row['score_away']:
                team_stats.loc[count, 'total_wins'] += 1
                team_stats.loc[count, 'home_wins'] += 1
            else:
                team_stats.loc[count, 'total_losses'] += 1           
        # loop through each row and looking for team in away and adding the score to our dataframe 
        elif row['team_away'] == team:
            team_stats.loc[count, 'total_points_scored'] += row['score_away']
            team_stats.loc[count, 'total_games_played'] += 1
            # Add each weather value that is not NaN to total temp to allow us to calculate avg temperature 
            if row['weather_temperature'] != 'NaN':
                team_stats.loc[count, 'total_temp'] += row['weather_temperature']
            else:
                continue
            # Calculate the number of wins and losses
            if row['score_away'] > row['score_home']:
                team_stats.loc[count, 'total_wins'] += 1
            else:
                team_stats.loc[count, 'total_losses'] += 1
        else:
            continue
    
    count += 1

# Add avg points scored column
team_stats['avg_points_scored'] = team_stats['total_points_scored'] / team_stats['total_games_played']
team_stats['avg_temperature'] = team_stats['total_temp'] / team_stats['total_games_played']
team_stats['win pct'] = (team_stats['total_wins'] / team_stats['total_games_played']) * 100
team_stats['Home win pct'] = (team_stats['home_wins'] / team_stats['home_games']) * 100
team_stats = team_stats[['team_name', 'total_points_scored', 'avg_points_scored', 'total_games_played', 'total_wins',
       'total_losses', 'avg_temperature', 'win pct', 'Home win pct']]
team_stats
```

```python
team_stats_sorted = team_stats.sort_values('avg_points_scored', ascending=False)
team_stats_sorted
```

```python
tiers = ['<10', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']
#create dataframe that is broken down by temperatures and displays avg over under, avg total_points scored in a game
weather_tiers = pd.DataFrame({
    'Tiers': tiers,
    'Total games': 0,
    'Total points': 0,
    'Total Over Under': 0,
    'Avg Total Pts/Game': 0,
    'Avg Over Under': 0
})
weather_tiers
```

```python
# Loop through each row of the merged data for 2002-2018
for index, row in merged_2002_2018.iterrows():
    # check the temperature for each row and place the total games, total points, and total over under into their respective temperature buckets
    if (row['weather_temperature'] < 10):
        count = 0
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 10) & (row['weather_temperature'] < 20):
        count = 1
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 20) & (row['weather_temperature'] < 30):
        count = 2
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 30) & (row['weather_temperature'] < 40):
        count = 3
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 40) & (row['weather_temperature'] < 50):
        count = 4
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 50) & (row['weather_temperature'] < 60):
        count = 5
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 60) & (row['weather_temperature'] < 70):
        count = 6
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 70) & (row['weather_temperature'] < 80):
        count = 7
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 80) & (row['weather_temperature'] < 90):
        count = 8
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    elif (row['weather_temperature'] >= 90):
        count = 9
        weather_tiers.loc[count, 'Total games'] += 1
        weather_tiers.loc[count, 'Total points'] += (row['score_home'] + row['score_away'])
        weather_tiers.loc[count, 'Total Over Under'] += row['over_under_line']
    else:
        continue
# Add Avg Total Pts per game and Avg Over Under to the weather_tiers DataFrame
weather_tiers['Avg Total Pts/Game'] = weather_tiers['Total points'] / weather_tiers['Total games']
weather_tiers['Avg Over Under'] = weather_tiers['Total Over Under'] / weather_tiers['Total games']
weather_tiers = weather_tiers[['Tiers', 'Total games', 'Avg Total Pts/Game', 'Avg Over Under']]
weather_tiers
```
